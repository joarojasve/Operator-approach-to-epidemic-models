```python
from scipy.optimize import fsolve
import math as ma
import numpy as np
from matplotlib import pylab as plt
I0=1
S0=9999
N=S0+I0
beta=0.2
gamma=0.2
BRN= beta/gamma
data=[I0,S0,BRN,N]
```

```python
def Sinf(x,data):
    return data[0]+data[1]-x+data[3]*ma.log(x/data[1])/data[2]
x0=1

xp=np.arange(0.05,5,0.05)
yp=[]
for r in xp:
    #print(r)
    data=[I0,S0,r,N]
    Vt = fsolve(Sinf, x0, args=data)
    #print(r, N-Vt[0])
    yp.append(N-Vt[0])
```

```python
plt.plot(xp,yp,label=r"$I(0)=1, \; N=10000$",color="navy")
plt.legend()
plt.xlabel(r"Basic reproductive number ($\mathcal{R}_0 $)")
plt.ylabel(r"Final size ($R_\infty$)")
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
```
